```python
import redis
# connecting to redis database
r=redis.Redis(host='localhost',port=6379,db=0)
# now we are loading the dataset
with open('food_data.json','r') as f:
  data=json.load(f) # loading json file data into the data variable
# storing the data in the redis
for key,value in data.items():
  r.set(key,json.dumps(value)) # here value are stored in dict format but json does not accept dict,
  # lists,set etc so json.dumps() coverts value sinto dict to string so the redis accept string
print("Data stored in Redis!")
```
