```python
# ==============================
# STEP 1: Upload kaggle.json
# ==============================
from google.colab import files as colab_files
colab_files.upload()

# ==============================
# STEP 2: Setup Kaggle
# ==============================
!pip install -q kaggle

!mkdir -p ~/.kaggle
!cp kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json

# ==============================
# STEP 3: Download Dataset
# ==============================
!kaggle datasets download -d harishkumardatalab/food-image-classification-dataset

# ==============================
# STEP 4: Unzip Dataset
# ==============================
!unzip -oq food-image-classification-dataset.zip -d data

# ==============================
# STEP 5: Import Libraries
# ==============================
import os
import shutil
import random
from zipfile import ZipFile

# ==============================
# STEP 6: Dataset Path
# ==============================
input_dir = "data/Food Classification dataset"

print("Classes Found:")
print(os.listdir(input_dir))

# ==============================
# STEP 7: Remove Old Processed Data
# ==============================
output_dir = "processed_data"

if os.path.exists(output_dir):
    shutil.rmtree(output_dir)

train_dir = os.path.join(output_dir, "train")
val_dir = os.path.join(output_dir, "val")

os.makedirs(train_dir)
os.makedirs(val_dir)

# ==============================
# STEP 8: Process Dataset
# Only use MAXIMUM 300 images/class
# Then apply 80-20 split
# ==============================

classes = os.listdir(input_dir)

for cls in classes:

    class_path = os.path.join(input_dir, cls)

    # Skip non-folder files
    if not os.path.isdir(class_path):
        continue

    # Get image files only
    image_list = [
        img for img in os.listdir(class_path)
        if img.lower().endswith((".jpg", ".jpeg", ".png"))
    ]

    # Shuffle images
    random.shuffle(image_list)

    # =====================================
    # LIMIT TO MAXIMUM 300 IMAGES ONLY
    # =====================================
    selected_images = image_list[:300]

    # =====================================
    # Dynamic 80-20 Split
    # =====================================
    split_index = int(len(selected_images) * 0.8)

    train_imgs = selected_images[:split_index]
    val_imgs = selected_images[split_index:]

    # Create folders
    os.makedirs(os.path.join(train_dir, cls))
    os.makedirs(os.path.join(val_dir, cls))

    # =====================================
    # Copy Train Images
    # =====================================
    for img in train_imgs:

        src_path = os.path.join(class_path, img)
        dst_path = os.path.join(train_dir, cls, img)

        shutil.copy(src_path, dst_path)

    # =====================================
    # Copy Validation Images
    # =====================================
    for img in val_imgs:

        src_path = os.path.join(class_path, img)
        dst_path = os.path.join(val_dir, cls, img)

        shutil.copy(src_path, dst_path)

    # Print Result
    print(f"{cls} → Total Used: {len(selected_images)} | Train: {len(train_imgs)} | Val: {len(val_imgs)}")

print("\ Dynamic dataset split completed successfully!")

# ==============================
# STEP 9: Create ZIP File
# ==============================
zip_path = "final_dataset.zip"

with ZipFile(zip_path, 'w') as zipf:

    for root, dirs, files in os.walk(output_dir):

        for file in files:

            file_path = os.path.join(root, file)

            zipf.write(
                file_path,
                os.path.relpath(file_path, output_dir)
            )

print(" Zip file created successfully!")

# ==============================
# STEP 10: Download ZIP
# ==============================
colab_files.download(zip_path)
```

```python
train_path="/content/processed_data/train" # train path
val_path="/content/processed_data/val" # val path
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    shear_range=0.2,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_data = train_datagen.flow_from_directory(
    train_path,
    target_size=(256,256),
    batch_size=32,
    class_mode='categorical'
)

val_data = val_datagen.flow_from_directory(
    val_path,
    target_size=(256,256),
    batch_size=32,
    class_mode='categorical'
)
```

```python
import warnings
warnings.filterwarnings('ignore')

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten

model = Sequential()

# Conv Layers
model.add(Conv2D(32, (3,3), activation='relu', input_shape=(256,256,3)))
model.add(MaxPool2D(2,2))

model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPool2D(2,2))

model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPool2D(2,2))

model.add(Flatten())

# Dense Layers
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))

#  FINAL LAYER (IMPORTANT)
model.add(Dense(34, activation='softmax'))  # 34 classes
```

```python
model.summary()
```

```python
# ==============================
# COMPILE MODEL
# ==============================
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# ==============================
# EARLY STOPPING
# ==============================
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

# ==============================
# TRAIN MODEL
# ==============================
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=20,
    callbacks=[early_stop]
)

# ==============================
# SAVE MODEL
# ==============================
model.save("food_classification_custom_model.h5")

print(" Model Saved Successfully!")

# ==============================
# EVALUATE MODEL
# ==============================
loss, accuracy = model.evaluate(val_data)

print(f"\nValidation Loss: {loss}")
print(f"Validation Accuracy: {accuracy*100:.2f}%")
```

```python
# ==============================
# LOAD IMAGE FOR PREDICTION
# ==============================
import numpy as np
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt

img_path = "/content/Crispy Chicken-Train (10).jpeg"  # change image path

img = image.load_img(img_path, target_size=(256,256))

img_array = image.img_to_array(img)

img_array = img_array / 255.0

img_array = np.expand_dims(img_array, axis=0)

# ==============================
# PREDICT
# ==============================
prediction = model.predict(img_array)

predicted_class = np.argmax(prediction)

# ==============================
# CLASS LABELS
# ==============================
class_labels = list(train_data.class_indices.keys())

result = class_labels[predicted_class]

# ==============================
# SHOW RESULT
# ==============================
plt.imshow(img)
plt.title(f"Predicted: {result}")
plt.axis("off")
plt.show()

print("Predicted Food:", result)
```

```python
train_path='/content/train'
val_path='/content/val'
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,Dense,MaxPool2D,Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
```

```python
#preprocessing image
train=ImageDataGenerator(rescale=1/255,rotation_range=20,shear_range=0.2,
                         zoom_range=0.2,horizontal_flip=True)
val=ImageDataGenerator(rescale=1/255)
```

```python
vgg16_train=train.flow_from_directory(train_path,target_size=(255,255),
                                      batch_size=32,class_mode='categorical')
vgg16_val=val.flow_from_directory(val_path,target_size=(255,255),
                                      batch_size=32,class_mode='categorical')
```

```python
from tensorflow.keras.applications import VGG16

vgg=VGG16(weights='imagenet',include_top=False,input_shape=(255,255,3))

for i in vgg.layers:
  i.trainable = False



```

```python
Flatten_values=Flatten()(vgg.output)

h1_out = Dense(units=128,kernel_initializer='he_uniform',activation='relu')(Flatten_values)
h2_out = Dense(units=64,kernel_initializer='he_uniform',activation='relu')(h1_out)
h3_out = Dense(units=32,kernel_initializer='he_uniform',activation='relu')(h2_out)
output = Dense(units=34,kernel_initializer='glorot_uniform',activation='softmax')(h3_out)

model = Model(inputs = vgg.input , outputs = output)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss = 'categorical_crossentropy',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)
```

```python
history = model.fit(
    vgg16_train,
    validation_data=vgg16_val,
    epochs=20,
    callbacks=[early_stop]
)
```
