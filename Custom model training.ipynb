```python
# ==============================
# STEP 1: Upload kaggle.json
# ==============================
from google.colab import files as colab_files
colab_files.upload()

# ==============================
# STEP 2: Setup Kaggle
# ==============================
!pip install -q kaggle

!mkdir -p ~/.kaggle
!cp kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json

# ==============================
# STEP 3: Download Dataset
# ==============================
!kaggle datasets download -d harishkumardatalab/food-image-classification-dataset

# ==============================
# STEP 4: Unzip Dataset
# ==============================
!unzip -oq food-image-classification-dataset.zip -d data

# ==============================
# STEP 5: Import Libraries
# ==============================
import os
import shutil
import random
from zipfile import ZipFile

# ==============================
# STEP 6: Dataset Path
# ==============================
input_dir = "data/Food Classification dataset"

print("Classes Found:")
print(os.listdir(input_dir))

# ==============================
# STEP 7: Remove Old Processed Data
# ==============================
output_dir = "processed_data"

if os.path.exists(output_dir):
    shutil.rmtree(output_dir)

train_dir = os.path.join(output_dir, "train")
val_dir = os.path.join(output_dir, "val")

os.makedirs(train_dir)
os.makedirs(val_dir)

# ==============================
# STEP 8: Process Dataset
# Only use MAXIMUM 300 images/class
# Then apply 80-20 split
# ==============================

classes = os.listdir(input_dir)

for cls in classes:

    class_path = os.path.join(input_dir, cls)

    # Skip non-folder files
    if not os.path.isdir(class_path):
        continue

    # Get image files only
    image_list = [
        img for img in os.listdir(class_path)
        if img.lower().endswith((".jpg", ".jpeg", ".png"))
    ]

    # Shuffle images
    random.shuffle(image_list)

    # =====================================
    # LIMIT TO MAXIMUM 300 IMAGES ONLY
    # =====================================
    selected_images = image_list[:300]

    # =====================================
    # Dynamic 80-20 Split
    # =====================================
    split_index = int(len(selected_images) * 0.8)

    train_imgs = selected_images[:split_index]
    val_imgs = selected_images[split_index:]

    # Create folders
    os.makedirs(os.path.join(train_dir, cls))
    os.makedirs(os.path.join(val_dir, cls))

    # =====================================
    # Copy Train Images
    # =====================================
    for img in train_imgs:

        src_path = os.path.join(class_path, img)
        dst_path = os.path.join(train_dir, cls, img)

        shutil.copy(src_path, dst_path)

    # =====================================
    # Copy Validation Images
    # =====================================
    for img in val_imgs:

        src_path = os.path.join(class_path, img)
        dst_path = os.path.join(val_dir, cls, img)

        shutil.copy(src_path, dst_path)

    # Print Result
    print(f"{cls} → Total Used: {len(selected_images)} | Train: {len(train_imgs)} | Val: {len(val_imgs)}")

print("\ Dynamic dataset split completed successfully!")

```

```python
train_path="/content/processed_data/train" # train path
val_path="/content/processed_data/val" # val path
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    shear_range=0.2,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_data = train_datagen.flow_from_directory(
    train_path,
    target_size=(255,255),
    batch_size=32,
    class_mode='categorical'
)

val_data = val_datagen.flow_from_directory(
    val_path,
    target_size=(255,255),
    batch_size=32,
    class_mode='categorical'
)
```

```python
import warnings
warnings.filterwarnings('ignore')

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten

model = Sequential()

# Conv Layers
model.add(Conv2D(32, (3,3), activation='relu', input_shape=(256,256,3)))
model.add(MaxPool2D(2,2))

model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPool2D(2,2))

model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPool2D(2,2))

model.add(Flatten())

# Dense Layers
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))

#  FINAL LAYER (IMPORTANT)
model.add(Dense(34, activation='softmax'))  # 34 classes
```

```python
model.summary()
```

```python
#model complile
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

#early stopping
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)


# Training model

history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10,
    callbacks=[early_stop]
)

# model saving
model.save("food_classification_custom_model.h5")

print(" Model Saved Successfully!")

#evaluate model

loss, accuracy = model.evaluate(val_data)

print(f"\nValidation Loss: {loss}")
print(f"Validation Accuracy: {accuracy*100:.2f}%")
```

```python
# model metrics

import tensorflow as tf
import numpy as np
import json

from tensorflow.keras.models import load_model

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    confusion_matrix,
    classification_report
)

# loading model successfully

model = load_model("food_classification_custom_model.h5")

print("Model Loaded Successfully")

# reset validation generator

val_data.reset()

# true labels

y_true = val_data.classes

# predictions

predictions = model.predict(val_data)

# predicted labels

y_pred = np.argmax(predictions, axis=1)

# accuracy

accuracy = accuracy_score(y_true, y_pred)

# precision

precision = precision_score(
    y_true,
    y_pred,
    average='weighted'
)

# confusion matrix

cm = confusion_matrix(y_true, y_pred)

# classification report

class_report = classification_report(
    y_true,
    y_pred,
    target_names=list(val_data.class_indices.keys()),
    output_dict=True
)

# store results

results = {
    "accuracy": float(accuracy),
    "precision": float(precision),
    "confusion_matrix": cm.tolist(),
    "classification_report": class_report
}

# save json file

with open("custom_cnn_metrics.json", "w") as json_file:

    json.dump(results, json_file, indent=4)

print("Metrics Saved Successfully!")
```
